# 1. Problema de Ingeniería (Ejemplo)

Un ingeniero agrícola necesita calcular el **volumen de agua** que puede almacenar un **tanque cilíndrico** para riego.

La fórmula del volumen de un cilindro es:

[
V = \pi r^2 h
]

donde:

* (r) = radio del tanque (m)
* (h) = altura del tanque (m)
* (V) = volumen (m³)

---

# 2. Definición del Algoritmo

## Entradas (Input)

Datos que el usuario debe ingresar:

* radio del tanque
* altura del tanque

Variables:

```
radio
altura
```

---

## Proceso (Cálculo)

Aplicar la fórmula del volumen del cilindro.

```
volumen = π * radio² * altura
```

---

## Salida (Output)

Mostrar el volumen del tanque.

```
Volumen del tanque en m³
```

---

# 3. Estructura IPO (Entrada – Proceso – Salida)

| Entrada | Proceso                       | Salida  |
| ------- | ----------------------------- | ------- |
| radio   | volumen = π * radio² * altura | volumen |
| altura  |                               |         |

---

# 4. Pseudocódigo

Antes de programar en Python, escribimos la lógica del algoritmo.

```
INICIO

leer radio
leer altura

volumen = π * radio^2 * altura

mostrar volumen

FIN
```

---

# 5. Código en Python

```python
# -------------------------------------------
# Programa: Cálculo del volumen de un tanque cilíndrico
# Autor: Estudiante de ingeniería
# -------------------------------------------

import math   # Biblioteca para usar el valor de pi

# Entrada de datos
radio = float(input("Ingrese el radio del tanque (m): "))
altura = float(input("Ingrese la altura del tanque (m): "))

# Proceso
volumen = math.pi * radio**2 * altura

# Salida
print("El volumen del tanque es:", volumen, "m³")
```
