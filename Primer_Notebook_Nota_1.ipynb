# 1. Problema de Ingeniería (Ejemplo)

Un ingeniero agrícola necesita calcular el **volumen de agua** que puede almacenar un **tanque cilíndrico** para riego.

La fórmula del volumen de un cilindro es:

[
V = \pi r^2 h
]

donde:

* (r) = radio del tanque (m)
* (h) = altura del tanque (m)
* (V) = volumen (m³)

---

# 2. Definición del Algoritmo

## Entradas (Input)

Datos que el usuario debe ingresar:

* radio del tanque
* altura del tanque

Variables:

```
radio
altura
```

---

## Proceso (Cálculo)

Aplicar la fórmula del volumen del cilindro.

```
volumen = π * radio² * altura
```

---

## Salida (Output)

Mostrar el volumen del tanque.

```
Volumen del tanque en m³
```

---

# 3. Estructura IPO (Entrada – Proceso – Salida)

| Entrada | Proceso                       | Salida  |
| ------- | ----------------------------- | ------- |
| radio   | volumen = π * radio² * altura | volumen |
| altura  |                               |         |

# 5. Código en Python

```python
# -------------------------------------------
# Programa: Cálculo del volumen de un tanque cilíndrico
# Autor: Estudiante de ingeniería
# -------------------------------------------

import math   # Biblioteca para usar el valor de pi

# Entrada de datos
radio = float(input("Ingrese el radio del tanque (m): "))
altura = float(input("Ingrese la altura del tanque (m): "))

# Proceso
volumen = math.pi * radio**2 * altura

# Salida
print("El volumen del tanque es:", volumen, "m³")
```

## Un ejemplo práctico: Cálculo de Caudal

En lugar de un simple "Hola Mundo", veamos algo aplicado. Supongamos que necesitas calcular el **caudal ($Q$)** que pasa por una tubería circular.

La fórmula física es:


$$Q = V \cdot A$$

Donde:

* $V$ es la velocidad del fluido.
* $A$ es el área de la sección transversal ($\pi \cdot r^2$).

### Estructura de pensamiento lógico

Antes de ver el código, analicemos los pasos que Python seguiría:

1. **Definir** el valor de la velocidad y el radio de la tubería.
2. **Calcular** el área del círculo.
3. **Multiplicar** velocidad por área para obtener el caudal.
4. **Mostrar** el resultado al usuario.

### Ejemplo de código en Python

Aquí tienes cómo se vería este proceso. Nota que uso nombres de variables claros, una buena práctica fundamental en ingeniería:

````
import math

# 1. Definición de variables con nombres descriptivos
radio_m = 0.5            # Radio en metros
velocidad_m_s = 2.0      # Velocidad en m/s

# 2. Cálculo del área (Usando la constante pi de la librería math)
area_m2 = math.pi * (radio_m ** 2)

# 3. Cálculo del caudal
caudal_m3_s = velocidad_m_s * area_m2

# 4. Mostrar el resultado
print(f"El caudal calculado es: {caudal_m3_s:.4f} m³/s")

````
