# **Dosificación de Mezcla para Columnas**

**Planteamiento:**

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos:**

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v=π*r^2*h$$

. (Sugerencia: usar 3.1416
 o importar math.pi).

4. Calcular la cantidad de sacos:
$$sacos=v*9.5$$

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3.

Sacos necesarios ≈3.58

 (el estudiante verá aquí la importancia de los tipos decimales).

```python
from __future__ import print_function
# ENTRADA: Solicitar la altura de la columna en metros.
# Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros:"))
d = float(input("Ingrese el diametro de la columna en metros:"))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d/2
v = 3.1416*r**2 *h
sacos = v*9.5

#SALIDA: Imprime el resultado
print("Volumen = ", round (v, 3),"m3.")
print("sacos necesarios =",round(sacos , 2))

```
