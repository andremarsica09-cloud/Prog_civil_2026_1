## 2. Situación problema

Un hogar recibe una cierta cantidad de **plumas de agua**.  
Se sabe que:

- **1 pluma = 0.025 litros por segundo**

Con esta información, el estudiante debe construir un programa que permita calcular:

1. el caudal total en litros por segundo;
2. el consumo diario en litros;
3. el consumo mensual en litros (suponiendo 30 días);
4. una clasificación del consumo:

````python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
# Se pide el número de plumas y se convierte a flotante para permitir valores decimales.
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
# Caudal de una pluma en litros por segundo
CAUDAL_PLUMA = 0.025  # litros/segundo
# Segundos en un día (24 horas * 60 minutos/hora * 60 segundos/minuto)
SEGUNDOS_DIA = 24 * 60 * 60 # segundos
# Días en un mes para el cálculo mensual
DIAS_MES = 30 # días

# 3. Calcular el caudal total
# Caudal total en litros por segundo
caudal_total = plumas * CAUDAL_PLUMA # litros/segundo

# 4. Calcular el consumo diario
# Consumo diario en litros
consumo_diario = caudal_total * SEGUNDOS_DIA # litros/día

# 5. Calcular el consumo mensual
# Consumo mensual en litros (suponiendo 30 días)
consumo_mensual = consumo_diario * DIAS_MES # litros/mes

# 6. Clasificar el consumo
# Determina el nivel de consumo (Bajo, Medio, Alto) basado en el consumo diario
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
# Presenta los resultados al usuario de manera clara
print(f"Número de plumas asignadas: {plumas}")
print(f"Caudal total: {caudal_total:.4f} litros/segundo")
print(f"Consumo diario: {consumo_diario:.2f} litros/día")
print(f"Consumo mensual (aprox. 30 días): {consumo_mensual:.2f} litros/mes")
print(f"Clasificación del consumo: {nivel_consumo}")
````

plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
# CAUDAL_PLUMA: Representa el caudal fijo de una "pluma de agua" en litros por segundo.
CAUDAL_PLUMA = 0.025  # litros/segundo
# SEGUNDOS_DIA: Representa el número total de segundos en un día completo (24 horas * 60 minutos/hora * 60 segundos/minuto).
SEGUNDOS_DIA = 24 * 60 * 60 # segundos

DIAS_MES = 30 # días

# Se calcula el caudal total que el hogar recibe multiplicando el número de plumas por el caudal unitario de una pluma.
caudal_total = plumas * CAUDAL_PLUMA # litros/segundo

# 4. Calcular el consumo diario
# Se calcula el consumo total de agua en un día. Esto se obtiene multiplicando el caudal total (litros/segundo)

consumo_diario = caudal_total * SEGUNDOS_DIA # litros/día

# 5. Calcular el consumo mensual
# Se estima el consumo total de agua en un mes, multiplicando el consumo diario por el número de días en el mes.
consumo_mensual = consumo_diario * DIAS_MES # litros/mes

# 7. Mostrar resultados

print(f"Caudal total: {caudal_total:.4f} litros/segundo")
print(f"Consumo diario: {consumo_diario:.2f} litros/día")
